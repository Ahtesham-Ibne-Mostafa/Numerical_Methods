P(x)= a3 + a2*x + a1*x^2 + a0*x^3

Then find approximate value of your created polynomial at x=3

Coefficients are given [a0,a1,a2,a3]=[2,1,4,3]

Q5. Suppose, you have 3 nodes(x0, x1, x2). We know standard format of Vandermonde matrix looks like:

[x0^0, x0^1, x0^2

 x1^0, x1^1, x1^2

 x2^0, x2^1, x2^2]

 But you have to create an alternate version of this matrix which should look like:

 [x2^2, x2^1, x2^0

  x1^2, x1^1, x1^0
  
  x0^2, x0^1, x0^0]

```python
#Q5
nodes=[1,2,3]
n=len(nodes)-1
V=np.zeros((n+1,n+1)) #fill up Dimension

#Use two nested loops to compute V matrix
for i in range(n+1):
    for j in range(n+1):
        V[i,j]=nodes[i]**j




print(V)

#Answer should look like:
#[[9,3,1]
# [4,2,1]
# [1,1,1]]
```

Q7. You all know the formula of Lagrange basis. But Messi created another basis which is called Lactane basis. the formula of this basis is given below:

You have to implement this new formula just like lagrange basis.

```python
#Q7
#node_x means an array of nodes, x means a value, t means a value
def lactane(t, node_x, x):
  L_t=1 #initialization
  #Use a loop to find basis
  x_t=node_x[t]
  for i in range(len(node_x)):
    if i != t:

        L_t += (x-node_x[i])/(node_x[t]-node_x[i])






  return L_t
```
